## ICUpatients5

```python
df_patients = pd.read_csv('./data/input/ICUpatients5.csv')
df_patients.head(3)
```

```python
data_types = {
    'subject_id': 'int64',
    'intime': 'datetime',
    'outtime': 'datetime',
}
```

```python
correct_format = '%Y-%m-%d %H:%M:%S UTC'

for col, dtype in data_types.items():
    if dtype == 'datetime':
        df_patients[col] = pd.to_datetime(df_patients[col], format=correct_format)
    else:
        df_patients[col] = df_patients[col].astype(dtype)
        
df_patients = df_patients[data_types.keys()]
```

```python
df_patients.loc[:, 'VMRegistro'] = 0
df_patients.loc[:, 'VMInicio'] = 0
df_patients.loc[:, 'VMReintubacao'] = 0
df_patients.loc[:, 'VMTempoPermanencia'] = 0
df_patients.loc[:, 'ProporcaoVMTempoPermanencia'] = 0
df_patients.loc[:, 'VMTempoDesmame'] = 0
```

```python
def get_registro_vm(row: pd.Series) -> int:
    
    if df_window.query("subject_id == @row.subject_id and registro == 1").shape[0]:
        return 1
    
    return 0

def get_vm_inicio(row: pd.Series) -> int:
    """
    Categorizar em faixas:
    0 - Sem registor de VM durante o período de internamento;
    1 - Primeiro registro de VM encontrado em até 24hrs (<=) a partir da data de internamento (intime no MIMIC);
    2 - Primeiro registro de VM encontrado entre 24hrs (>) e 48hrs (<=) a partir da data de internamento (intime no MIMIC);
    3 - Primeiro registro de VM encontrado entre 48hrs (>) e 72hrs (<=) a partir da data de internamento (intime no MIMIC);
    4 - Primeiro registro de VM encontrado entre 72hrs (>) a partir da data de internamento (intime no MIMIC).
    """
    
    index = df_window.query("subject_id == @row.subject_id and registro == 1").sort_values('day', ascending=True).index
    
    for ix in index:
        if day := df_window.loc[ix].day:
            if day <= 1:
                return 1
            elif day <= 2:
                return 2
            elif day <= 3:
                return 3
            else:
                return 4
    
    return 0

def get_vm_reintubacao(row: pd.Series) -> int:
    """
    No MIMIC-IV significa que o paciente apresenta mais de 1 registro de VM onde ele sai da ventilação mecânica (endtime) e entra novamente após 72hrs. (start_time do registro imediatamente subsequente).
    Valores possíveis:
    0 - Não houve registro de re-intubação durante o período de internamento;
    1 - Houve registro de re-intubação durante o período de internamento.
    """
    
    index = df_window.query("subject_id == @row.subject_id and registro == 1").sort_values('day', ascending=True).index
    
    # Difference == 1, significa que o paciente continuou na ventilação
    # Difference == 2, significa que o paciente saiu da ventilação e entrou novamente até 24hrs
    # Difference == 3, significa que o paciente saiu da ventilação e entrou novamente 24hrs < x <= 48hrs
    # Difference == 4, significa que o paciente saiu da ventilação e entrou novamente 48hrs < x <= 72hrs
    # Difference > 4, significa que o paciente saiu da ventilação e entrou novamente após mais de 72hrs
    
    if not index.shape[0]:
        return 0
    
    for i in range(1, index.shape[0]):
        day_1 = df_window.loc[index[i - 1], 'day']
        day_2 = df_window.loc[index[i], 'day']
        
        difference = np.abs(day_1 - day_2)
        if difference > 4:
            return 1

    return 0

def get_vm_tempo_permanencia(row: pd.Series) -> int:
    """
    Gerar mapa diária de janelas de 24hrs com algum registro de VM.
    Obs:. Quando start_time estiver dentro de 1 janela de 24hrs e end_time em outra, registrar uso de VM na janela onde o start_time estiver enquadrado, na janela onde o end_time estiver enquadrado e em todas as janelas entre uma e outra.
    Quando houver episódios de re-intubação, somar a quantidade de dias em que o paciente fica entubado em cada episódio.
    Categorizar em faixas:
    0 - Não teve uso ou não foi encontrada anotação;
    1 - 0 < x <= 25%;
    2 - 25% < x <= 50%;
    3 - 50% < x <= 75%;
    4 - 75% < x <= 100%.
    """
    index = df_window.query("subject_id == @row.subject_id and registro == 1").sort_values('day', ascending=True).index
    days_on_icu = df_window[df_window.subject_id == row.subject_id].index.shape[0]
    
    # Difference == 1, significa que o paciente continuou na ventilação
    # Difference == 2, significa que o paciente saiu da ventilação e entrou novamente até 24hrs
    # Difference == 3, significa que o paciente saiu da ventilação e entrou novamente 24hrs < x <= 48hrs
    # Difference == 4, significa que o paciente saiu da ventilação e entrou novamente 48hrs < x <= 72hrs
    # Difference > 4, significa que o paciente saiu da ventilação e entrou novamente após mais de 72hrs
    
    if not index.shape[0]:
        return 0, 0

    start_window = [index[0]]
    end_window = []

    for i in range(1, index.shape[0]):
        day_1 = df_window.loc[index[i - 1], 'day']
        day_2 = df_window.loc[index[i], 'day']
        
        difference = np.abs(day_1 - day_2)
        if difference > 4:
            start_window.append(index[i])
            end_window.append(index[i-1])

    end_window.append(index[-1])

    registros_vm = 0
    for start, end in zip(start_window, end_window):
        start_day = df_window.loc[start, 'day']
        end_day = df_window.loc[end, 'day']
        
        difference = np.abs(start_day - end_day) + 1

        registros_vm += difference
        
    proportion = registros_vm/days_on_icu
    
    if 0 < proportion <= 0.25:
        return 1, proportion
    
    if 0.25 < proportion <= 0.50:
        return 2, proportion
    
    if 0.50 < proportion <= 0.75:
        return 3, proportion
    
    if 0.75 < proportion <= 1:
        return 4, proportion
    
    return -1, None

def get_vm_tempo_desmame(row: pd.Series) -> int:
    """
    Para cada episódio de intubação/re-intubação, contar a quantidade de dias desde o primeiro dia que ele sai da ventilação mecânica até o último dia de registro da ventilação mecânica do episódio.
    Depois somar a quantidade de dias de desmame obtida em cada episódio.
    Se o paciente sai e não volta ou re-intuba, enquadrar na faixa "até 24hrs".
    Categorizar em faixas:
    0 - Não teve uso ou não foi encotrada anotação;
    1 - Até 24hrs (<=);
    2 - Entre 24hrs (>) e 48hrs (<=);
    3 - Entre 48hrs (>) e 72hrs (<=);
    4 - Acima de 72hrs (>).
    """
    index = df_window.query("subject_id == @row.subject_id and registro == 1").sort_values('day', ascending=True).index
    
    if not index.shape[0]:
        return 0
    
    # Difference == 1, significa que o paciente continuou na ventilação
    # Difference == 2, significa que o paciente saiu da ventilação e entrou novamente até 24hrs
    # Difference == 3, significa que o paciente saiu da ventilação e entrou novamente 24hrs < x <= 48hrs
    # Difference == 4, significa que o paciente saiu da ventilação e entrou novamente 48hrs < x <= 72hrs
    # Difference > 4, significa que o paciente saiu da ventilação e entrou novamente após mais de 72hrs
    
    if index.shape[0] == 1:
        return 1

    start_window = [index[0]]
    end_window = []
    for i in range(1, index.shape[0]):
        day_1 = df_window.loc[index[i - 1], 'day']
        day_2 = df_window.loc[index[i], 'day']
        
        difference = np.abs(day_1 - day_2)
        if difference > 4:
            start_window.append(index[i])
            end_window.append(index[i-1])

    end_window.append(index[-1])

    contador = 0
    for start, end in zip(start_window, end_window):
        for i in range(start, end+1):
            if not df_window.loc[i, 'registro']:
                registro_final = df_window.loc[end, 'day']
                registro_atual = df_window.loc[i, 'day']
                
                contador += (np.abs(registro_final - registro_atual) + 1)
                break

    contador += len(start_window)
        
    if contador == 1:
        return 1
    if contador == 2:
        return 2
    elif contador == 3:
        return 3
    elif contador > 4:
        return 4

    return 1
```

```python
df_patients.loc[:, 'VMRegistro'] = df_patients.apply(get_registro_vm, axis=1)
```

```python
result = df_patients.apply(get_vm_tempo_permanencia, axis=1)
df_patients.loc[:, 'VMTempoPermanencia'] = [x[0] for x in result]
df_patients.loc[:, 'ProporcaoVMTempoPermanencia'] = [x[1] for x in result]
```

```python
df_patients.loc[:, 'VMTempoDesmame'] = df_patients.apply(get_vm_tempo_desmame, axis=1)
```
