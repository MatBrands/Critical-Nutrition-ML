```python
import pandas as pd

from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
```

## PHASE 1

```python
df = pd.read_csv('data/PHASE1_PATIENTS_SAMPLE.csv')
df.head()
```

```python
for column in df.select_dtypes(include='object'):
    df[column] = df[column].astype('category').cat.codes
```

```python
features = df.drop(columns = 'denouement')
target = df.denouement
```

```python
def create_train(features: pd.DataFrame, target: pd.Series) -> tuple:
    x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42, stratify = target)
    
    x_smote, y_smote = x_train.drop(columns = 'subject_id'), y_train
    
    x_smote, y_smote = SMOTE(random_state = 42).fit_resample(x_smote, y_smote)
    
    pd.concat([pd.DataFrame(x_smote), pd.DataFrame(y_smote)], axis = 1).to_csv('data/TRAIN_SMOTE.csv', index = False)
    pd.concat([pd.DataFrame(x_train), pd.DataFrame(y_train)], axis = 1).to_csv('data/TRAIN.csv', index = False)
    pd.concat([pd.DataFrame(x_test), pd.DataFrame(y_test)], axis = 1).to_csv('data/TEST.csv', index = False)
```

```python
create_train(features, target)
```
