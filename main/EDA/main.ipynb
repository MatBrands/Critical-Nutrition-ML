```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/PHASE1_PATIENTS_SAMPLE.csv')
df.drop('subject_id', axis=1, inplace=True)
```

```python
# Selecionar as colunas com mais de 5 valores unicos

df_numerical = df.loc[:, df.apply(lambda x: x.nunique() > 6)]
df_numerical = pd.concat([df.denouement, df_numerical], axis=1)
```

```python
def plot_correlation_matrix(df: pd, method, ax) -> list:
    corr = df.corr(numeric_only=True, method=method)
    sns.heatmap(corr, cmap="crest", vmax=.3, center=0,
        square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt='.2f', ax=ax)
    ax.set_title(f'MÃ©todo: {method.capitalize()}', fontsize='large', fontweight='bold', style='italic', family='monospace')
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))

plot_correlation_matrix(df_numerical, 'pearson', ax)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))

plot_correlation_matrix(df_numerical, 'kendall', ax)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))

plot_correlation_matrix(df_numerical, 'spearman', ax)

plt.show()
```
